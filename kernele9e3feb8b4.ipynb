```python
#Selecting rows with no missing year, price  or area
cdf = df[df['YR_RMDL'].notnull() & df['PRICE'].notnull()&  df['LANDAREA'].notnull()]
cdf
```

```python
cdf.count()
```

```python
def price_per_area(price, area):
    return price / area

cdf['price_per_area'] = price_per_area(df['PRICE'], df['LANDAREA'])
cdf.head()
```

```python
cdf.groupby('YR_RMDL')['price_per_area'].mean()
```

```python
#Standard Deviation Method

meanppa = cdf['price_per_area'].mean()
stdppa  = cdf['price_per_area'].std()
toprange = meanppa + stdppa * 1.96
botrange = meanppa - stdppa * 1.96

copydf = cdf.copy() #to not mess up the original df
copydf = copydf.drop(copydf[copydf['price_per_area'] > toprange].index)
copydf = copydf.drop(copydf[copydf['price_per_area'] < botrange].index)

copydf.head()
```

```python
copydf.groupby('YR_RMDL')['price_per_area'].mean()
```

```python
#Define bins 
bins = [1880, 1900, 1920, 1940, 1960, 1980, 2000, 2020]

# Create names for the four groups
group_names = ['1880-1899', '1900-1919', '1920-1939', '1940-1959', '1960-1979', '1980-1999', '2000-2019']

#make new column with letter grades
copydf['group'] = pd.cut(copydf['YR_RMDL'], bins, labels=group_names)
copydf.head()
```

```python
copydf.groupby('group')['price_per_area'].mean()
```
